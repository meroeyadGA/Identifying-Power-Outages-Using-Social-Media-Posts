```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Import weather data from year 2014 - 2019 for 8 states
data_list = []
for i in range(1, 9):
    data = pd.read_csv(f"../Data/WeatherData_2014_2019_State_{i}.csv", encoding = 'unicode_escape', index_col=None, header=0)
    data_list.append(data)
df = pd.concat(data_list, axis=0, ignore_index=True)
```
