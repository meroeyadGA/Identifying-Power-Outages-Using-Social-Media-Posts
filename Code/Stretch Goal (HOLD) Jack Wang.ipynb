## Merge the two dataframes 

```python
twitter_df['date_only'] = [f"{post.split('|')[0].split()[0]}" 
 for i, post in enumerate(twitter_df['time_location'])]

power_df['date_only'] = [f"{post.split('|')[0].split()[0]}" 
 for i, post in enumerate(power_df['time_location'])]
```

```python
# power_df = power_df.drop(columns=['Unnamed: 0', 'Month', 'Alert Criteria', 'Number of Customers Affected',
#                                  'Hour Event Began', 'NERC Region', 'Alert Criteria', 'Area Affected',
#                                  'Demand Loss (MW)'])
#df.drop(columns=['tweet_word_count', 'user_name', 'port_tweet', 'num_likes', 'num_retweets', 'Unnamed: 0', 'Month', 'NERC Region', 'Hour Event Began' ])
```

```python
df = pd.merge(twitter_df, power_df, on='date_only', how='inner')
df.shape
```

```python
df_final = df.loc[:,['tweet', 'location', 'date_only', 'state', 'Number of Customers Affected', 'Demand Loss (MW)']]
```
