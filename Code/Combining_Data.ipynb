```python
t0 = time.time()
twitter_df['target'] = 0
size = len(twitter_df['location'].unique())
counter = 0

# we need to check the locations and times match up so let's make smaller dataframes and check those
for place in twitter_df['location'].unique():
    
    # check Unknown states as well
    state_power_df = power_df[(power_df['state'] == place) | (power_df['state'] == 'Unknown')]
    state_twitter_df = twitter_df[(twitter_df['location'] == place)]
    
    # loop through every tweet in the state twitter df
    for i, tweet_ts in enumerate(state_twitter_df['time_stamp']):
        
        # convert tweet timestamp to datetime format
        tweet_time = datetime.strptime(tweet_ts, '%Y-%m-%d %H:%M:%S')
        
        # loop through every outage in the state power df
        for j, start_ts in enumerate(state_power_df['Date Event Began']):
            
            # convert the start time to datetime format
            event_start_str = start_ts + " " + state_power_df['Time Event Began'][state_power_df['state'].index[j]]
            start_time = datetime.strptime(event_start_str, '%Y-%m-%d %H:%M:%S')

            # convert the restoration time to dattime format
            resto_str = state_power_df['Date of Restoration'][state_power_df['state'].index[j]] + " " + state_power_df['Time of Restoration'][state_power_df['state'].index[j]]

            # we need this try/except because some Dates and Times of resotration are unknowns, so we're going to 
            # set resto_time to be equal to one day after start time (may need to change this)
            try:
                resto_time = datetime.strptime(resto_str, '%m/%d/%Y %H:%M:%S')
            except:
                resto_time = start_time + timedelta(days=1)

            # check if tweet_time is in between start_time and resto_time
            if (tweet_time >= start_time) and (tweet_time <= resto_time):
                twitter_df['target'][state_twitter_df['location'].index[i]] = 1
    
    if (round((counter / size)*100) % 5 == 0) and round((counter / size)*100) > 0:
        print(f'{round((counter / size)*100)}% complete')
        print(f'{(time.time() - t0)/60} mins')
        print(f"{twitter_df['target'].sum()} total hits")
        print('---------------------')
    counter += 1
    
print('100% complete')
print(f'{(time.time() - t0)/60} mins')
print(f"{twitter_df['target'].sum()} total hits")
```
